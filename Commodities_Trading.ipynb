### Question 4
Check the stationarity of the `Heating Oil - Brent` spread. What are your thoughts on the results?

To check the stationarity, we are going to use an Augmented Dickey Fuller test:
- The more negative the ADF statistic, the higher the likelihood of the time series being stationary.

- The p-value serves as a measure of the strength of evidence against the null hypothesis, which posits that the time series is non-stationary. When the p-value falls below a critical threshold, we will choose 0.05 here , we reject the null hypothesis, indicating that the time series is stationary. 

- Critical values establish benchmarks for null hypothesis rejection at significance levels of 1%, 5%, and 10%. If the ADF statistic is lower than the critical value corresponding to a specific significance level, we reject the null hypothesis, indicating stationarity at that significance level.

```python
#CODE
from statsmodels.tsa.stattools import adfuller

# Compute the ADF test for the spread
adf_result = adfuller(df_concat["Heating Oil - Brent Spread"])

# Extract and print the results
adf_statistic = adf_result[0]
p_value = adf_result[1]
critical_values = adf_result[4]

print("\nADF Statistic:", adf_statistic)
print("\np-value:", p_value)
print("\nCritical Values:")
for key, value in critical_values.items():
    print(f"\t{key}: {value}")
```

### Interpretation of results:
- Our ADF Statistic is greater than the critical values at 1%, 5% and 10%
- The p-value is greater than 0.05

Therefore, we fail to reject the null hypothesis and we conclude that the spread is likely non-stationary

### To study the seasonality, we are going to use seasonal_decompose

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform seasonal decomposition
decomposition = seasonal_decompose(df['Weekly U.S. Product Supplied of Finished Motor Gasoline  (Thousand Barrels per Day)'],
                                    model='multiplicative', period=52)  # Assuming weekly data with a yearly seasonality

# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df, label='Original', color='skyblue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color="pink")
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal', color="palegreen")
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual', color='salmon')
plt.legend(loc='upper left')
plt.tight_layout()

# Remove the seasonal component
gasoline_demand_seasonality_removed = df['Weekly U.S. Product Supplied of Finished Motor Gasoline  (Thousand Barrels per Day)']/decomposition.seasonal.dropna()

# Plot the original data and the data with seasonality removed
plt.figure(figsize=(12, 8))
plt.plot(df.index, df,color='indigo',label='Original')
plt.plot(gasoline_demand_seasonality_removed.index, gasoline_demand_seasonality_removed,color='peachpuff',label='Seasonality Removed')
plt.legend(loc='upper left')
plt.title('Original Data vs Data with Seasonality Removed')
plt.show()
```

### Let's compute the price elasticity

We know that PED = $-\frac{dD}{dP} \times \frac{P}{D}$
where D is the demand and P is the price. 

To calculate the price-elasticity of the demand we do a univariate OLS where y is the demand and x is the price such that $y = \beta \cdot x + \alpha$ . 

Then $\frac{{dy}}{{dx}} = \beta$ and we are able to compute the price-elasticity of the demand.

- We resample the price dataframe to have the same format than the demand dataframe

```python
df_resample=df_concat['US Gasoline Prices'].resample('W-Fri').last().iloc[:-2]
df_resample
```

- We see that the values start at the date 2010-01-08 so we need to limit the demand to this period

```python
gasoline_demand_seasonality_removed_resample=gasoline_demand_seasonality_removed["2010-01-08":]
gasoline_demand_seasonality_removed_resample
```

### We check that we have the same shape for the two series

```python
gasoline_demand_seasonality_removed_resample.shape==df_resample.shape
```

- We do the univariate OLS

```python
# Import the statsmodels library
import statsmodels.api as sm

# Assign X and Y variables
X = df_resample
Y = gasoline_demand_seasonality_removed_resample

# Add constant term to the independent variable
X = sm.add_constant(X) 

# Create an Ordinary Least Squares (OLS) regression model
model = sm.OLS(Y,X)

# Fit the OLS model to the data
results = model.fit()

# Extract parameters from the fitted model
alpha = results.params.values[0]
beta = results.params.values[1]

# Calculate price elasticity using the estimated parameters
price_elasticity = - beta * (np.mean(X['US Gasoline Prices']) / np.mean(Y))

# Print the calculated price elasticity of demand for gasoline
print("Price-Elasticity of Demand for Gasoline:", price_elasticity)
```

### Computation of the historical margin:

The costs are:
- Operating costs
- The cost of buying Brend oil

The revenues come from:
- 65% of Gasoline
- 35% of Distilates

```python
def compute_margin(gasoline_prices, distilate_prices, brent_prices, operating_costs, yield_gasoline, yield_distilate):
    Margin=yield_gasoline*gasoline_prices + yield_distilate*distilate_prices - operating_costs - brent_prices
    return Margin

operating_costs=6
gasoline_prices=df_concat['US Gasoline Prices']
distilate_prices=df_concat['Heating Oil Prices']
brent_prices=df_concat['Brent Prices']
yield_gasoline=0.65
yield_distilate=0.35
Margin=compute_margin(gasoline_prices, distilate_prices, brent_prices, operating_costs, yield_gasoline, yield_distilate)
Margin
```

### Question 10

Using a 5% discounting rate, assuming that our refinery will produce until Dec 2027, and has a capacity of 100'000 barrels per day, how would you evaluate the asset?

```python
def compute_npv(start_date,end_date, future_margin, discount_rate, capacity):
    data_prices=future_margin.loc[start_date:end_date]
    monthly_revenue = data_prices * capacity*30
    time_diff_years = (monthly_revenue.index - pd.Timestamp.now()) / pd.Timedelta(days=365)
    # Calculate the discount factors
    discount_factors = 1 / ((1 + discount_rate) ** time_diff_years)
    # Compute the discounted cash flows
    discounted_cash_flows = monthly_revenue * discount_factors
    # Compute the NPV by summing all discounted cash flows
    npv = discounted_cash_flows.sum()
    return npv
start_date="2024-06-30"
end_date="2027-12-31"
future_margin=Margin_future
discount_rate = 0.05  # Discount rate (5%)
capacity = 100000
npv=compute_npv(start_date,end_date, future_margin, discount_rate, capacity)
print(f"Using a 5% discounting rate, assuming that our refinery will produce until Dec 2027, and has a capacity of 100'000 barrels per day, we find that the value of the asset is: ${npv:,.2f}")
```
