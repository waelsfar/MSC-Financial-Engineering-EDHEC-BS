# Importation of Data from the S&P500

```python
URL = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies' # Wikipedia Webpage

L_tickers = pd.read_html(URL)[0]['Symbol'].tolist() #List Containing the Ticker names

# Define start and end dates
start_date = '2000-01-01'
end_date = '2022-12-31'

# Fetch adjusted close prices for the selected stocks
stock_data = yf.download(L_tickers, start=start_date, end=end_date)['Adj Close']


stock_data
```

# Calculation of Volatility

```python
# Here we add a columns Year in our dataframe
L_year =[]
L_month=[]
for date in return_df.index:
    L_year.append(date.year)
    L_month.append(date.month)
return_df['Year']=L_year
return_df['Month']=L_month
return_df['Month-Year']=pd.to_datetime(return_df[['Month','Year']].assign(DAY=1))
return_df
```

```python
L_date_unique=list(np.unique(return_df['Month-Year']))

df_monthly_volatility=pd.DataFrame(index=L_date_unique, columns=return_df.columns[:363])

for stock in return_df.columns[:363]:
    L_mean=[]
    L_std=[]
    L_sr=[]
    for date in L_date_unique:
        data= return_df[return_df['Month-Year']== date][stock] 
        std=data.std()*(21**0.5) # Volatility
        L_std.append(std)

    df_monthly_volatility[stock]=L_std

df_monthly_volatility # This dataframe will be composed of all the volatility per year and per stock
```

# Calculation of Metrics

```python
def mean_squared_error(y_true, y_pred):

    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length")

    # Calculate the squared differences between true and predicted values
    squared_errors = [(true - pred) ** 2 for true, pred in zip(y_true, y_pred)]

    # Calculate the mean squared error
    mse = sum(squared_errors) / len(y_true)

    return mse
```

```python
def sharpe_ratio(returns, risk_free_rate):

    mean_return = np.mean(returns)
    std_return = np.std(returns)

    # Calculate the excess return
    excess_return = mean_return - risk_free_rate

    # Calculate the Sharpe Ratio
    sharpe_ratio = excess_return / std_return

    return sharpe_ratio
```

```python
def information_ratio(portfolio_returns, benchmark_returns):

    # Calculate the active return
    active_return = np.array(portfolio_returns) - np.array(benchmark_returns)

    # Calculate the tracking error (standard deviation of active return)
    tracking_error = np.std(active_return)

    # Calculate the Information Ratio
    information_ratio = np.mean(active_return) / tracking_error

    return information_ratio
```

```python
def calculate_skewness(data):
 
    n = len(data)
    mean_data = np.mean(data)
    std_data = np.std(data)

    skewness = (n / ((n - 1) * (n - 2))) * np.sum(((data - mean_data) / std_data) ** 3)

    return skewness
```

```python
def calculate_maximum_drawdown(prices):

    peak_value = np.maximum.accumulate(prices)
    trough_value = np.minimum.accumulate(prices)
    
    drawdowns = (peak_value - trough_value) / peak_value
    max_drawdown = np.max(drawdowns) * 100  # Convert to percentage

    return max_drawdown
```

```python
def calculate_hit_ratio(profitable_trades, total_trades):

    if total_trades == 0:
        return 0.0

    hit_ratio = (profitable_trades / total_trades) * 100
    return hit_ratio
```

```python
def calculate_turnover(buy_value, sell_value, average_portfolio_value):

    total_value_traded = buy_value + sell_value
    turnover_ratio = (total_value_traded / average_portfolio_value) * 100
    return turnover_ratio
```
