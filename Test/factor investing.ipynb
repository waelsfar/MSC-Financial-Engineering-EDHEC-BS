# Optimized Portfolio

```python
from scipy.optimize import minimize
```

### We start by calculating the initial carbon intensity

```python
portfolio_carbon_intensity_initial=round(np.sum(df['Carbon intensity']*df['Weight']),2)
portfolio_carbon_intensity_initial
```

### Then we compute the targeted carbon intensity

```python
portfolio_carbon_intensity_targeted=round(portfolio_carbon_intensity_initial*0.7,2)
portfolio_carbon_intensity_targeted
```

### We compute the minimum and maximum weights that are accepted

```python
df['Min Weight']=0.5*df['Weight']
df['Max Weight']=1.2*df['Weight']
df
```

### We create a dataframe with only the weights for a better visibility

```python
df_weight=pd.DataFrame()
df_weight["Name"]=df["Name"]
df_weight["Initial Weight"]=df["Weight"]
df_weight["Min Weight"]=df["Min Weight"]
df_weight["Max Weight"]=df["Max Weight"]
df_weight
```

### We set the maximum deviation for buy and sell movements

```python
max_deviation = 0.2
```

### Optimization

```python
# Data we will use:

initial_weights = np.array(df["Weight"])
carbon_intensity = np.array(df["Carbon intensity"])
net_income = np.array(df["NET_INCOME"])
max_deviation = 0.2
# Calculate the initial portfolio's carbon intensity:

initial_portfolio_intensity = round(np.sum(carbon_intensity*initial_weights),2)

# Target carbon intensity (in our case, we want to reduce by 30%):

target_intensity = round(initial_portfolio_intensity * 0.7,2)

# Constraint on sum of the weights:

def constraint1(weights):
    return np.sum(weights) - 100

# Constraint on min weight:

def constraint2(weights):
    return weights - (initial_weights * 0.5)

# Constraint on max weight:

def constraint3(weights):
    return (initial_weights * 1.2) - weights

# Constraint on carbon intensity:

def constraint4(weights):
    return target_intensity - np.dot(weights, carbon_intensity)

# Constraint on max deviation:

def constraint5(weights):
    max_deviation = np.sum(initial_weights) * 0.2
    return np.sum(np.abs(weights - initial_weights)) - max_deviation

# Define objective function (find the weights):

def objective_function(weights):
    return np.sum(np.abs(weights - initial_weights))

# Define bounds
bounds = [(0, 100)] * len(initial_weights)

# Define constraints (eq=equation & ineq=inequation)
constraints = [
    {'type': 'eq', 'fun': constraint1},
    {'type': 'ineq', 'fun': constraint2},
    {'type': 'ineq', 'fun': constraint3},
    {'type': 'eq', 'fun': constraint4},
    {'type': 'ineq', 'fun': constraint5}
]

# Initial guess for weights
initial_guess = initial_weights

# Run optimization
result = minimize(objective_function, initial_guess, bounds=bounds, constraints=constraints)

# Get optimized weights
optimized_weights = result.x
df_weight["Optimized Weight"]=optimized_weights
# Calculate the carbon intensity of the new portfolio
new_portfolio_intensity = round(np.sum(carbon_intensity*optimized_weights),2)

# Print results
print("Initial Portfolio Carbon Intensity:", initial_portfolio_intensity)
print("Target Portfolio Carbon Intensity:", target_intensity)
print("New Portfolio Carbon Intensity:", new_portfolio_intensity)
```
