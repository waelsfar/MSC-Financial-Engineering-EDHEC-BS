Now use a rolling window of T
est = 60 months to estimate the portfolio
weights for the two strategies listed above for each of the T − T
est
months. That is, repeat the calculations of the previous question for all
the dates after the first 60 months.

```python
def gmv(matrix):
    l = len(matrix)
    vector_ones = np.ones(l)
    matrix_inv = np.linalg.inv(matrix)
    numerator = np.dot(matrix_inv, vector_ones)
    weights = numerator / np.sum(numerator)
    return weights

T_est = 60
gmv_weights_list = []

for i in range(T_est, len(monthly_returns)):
    returns_window = monthly_returns.iloc[i - T_est:i]
    covariance_matrix = returns_window.cov()
    gmv_weights = gmv(covariance_matrix.values)
    gmv_weights_list.append(gmv_weights)

gmv_weights_df = pd.DataFrame(gmv_weights_list, columns=monthly_returns.columns)

print("GMV Portfolio Weights:")
print(gmv_weights_df)
```
