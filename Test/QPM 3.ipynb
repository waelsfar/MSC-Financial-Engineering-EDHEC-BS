```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies' # Wikipedia Webpage

L_tickers = pd.read_html(URL)[0]['Symbol'].tolist() #List Containing the Ticker names

# Define start and end dates
start_date = '2000-01-01'
end_date = '2022-12-31'

# Fetch adjusted close prices for the selected stocks
stock_data = yf.download(L_tickers, start=start_date, end=end_date)['Adj Close']


stock_data
```

- Drop the columns that have only “NaN” entries.

```python
stocks_prices=stock_data
```

```python
stocks_prices.dropna(axis=1, how='all', inplace=True)
```

```python
stocks_prices
```

-Drop also the company names that have more than 100 missing observations.

```python
for stock in stocks_prices.columns:
    s=stocks_prices[stock].isna().sum() # Count the number of NaN
    if s>100:
        stocks_prices=stocks_prices.drop(columns=stock)
stocks_prices
```

# Question 2

- Compute the log returns for the companies in your dataset.

```python
log_return_df=np.log(stocks_prices / stocks_prices.shift())
```

```python
log_return_df
```

# Question 3

- Compute the annual mean return, volatility, and Sharpe ratios for these companies in your dataset.

```python
# Here we add a columns Year in our dataframe
L_year =[]
for date in log_return_df.index:
    L_year.append(date.year)

log_return_df['Year']=L_year
log_return_df
```

```python
L_year_unique=list(np.unique(L_year))

df_annual_mean_return=pd.DataFrame(index=L_year_unique, columns=log_return_df.columns[:364])
df_annual_volatility=pd.DataFrame(index=L_year_unique, columns=log_return_df.columns[:364])
df_annual_sr=pd.DataFrame(index=L_year_unique, columns=log_return_df.columns[:364])

for stock in log_return_df.columns[:364]:
    L_mean=[]
    L_std=[]
    L_sr=[]
    for year in L_year_unique:
        data= log_return_df[log_return_df['Year']== year][stock] 
        m=data.mean()*252 # Mean
        std=data.std()*(252**0.5) # Volatility
        sr=m/std # Sharpe Ratio We make the hypothesis that the risk free rate is 0
        
        L_mean.append(m)
        L_std.append(std)
        L_sr.append(sr)
        
    df_annual_mean_return[stock]=L_mean
    df_annual_volatility[stock]=L_std
    df_annual_sr[stock]=L_sr
```

# Question 4:

- Would it make sense to choose portfolio weights based only on the Sharpe ratios of the stocks in your dataset? Explain the reasons for your answer

To answer this question let calculate the skewness and the kurtosis

```python
from scipy.stats import skew, kurtosis


df_skew=pd.DataFrame(index=L_year_unique, columns=log_return_df.columns[:364])
df_kurtosis=pd.DataFrame(index=L_year_unique, columns=log_return_df.columns[:364])


for stock in log_return_df.columns[:364]:
    L_skew=[]
    L_kurtosis=[]

    for year in L_year_unique:
        data= log_return_df[log_return_df['Year']== year][stock]
        skewness=skew(data)
        kurt=kurtosis(data)
       
        L_skew.append(skewness)
        L_kurtosis.append(kurt)

    df_skew[stock]=L_skew
    df_kurtosis[stock]=L_kurtosis



```

# Sharpe Ratio Limitations:


- The Sharpe ratio has several limitations:

- Sharpe ratio is appropriate for efficient portfolios but not for individual assets because it ignores correlation between these assets.
- Sharpe ratio, because it depends on the mean and variance of returns, is not appropriate if returns have high skewness and kurtosis.
- Sharpe ratio looks at first two moments, but ignores the distribution of returns. In contrast, the Omega measure uses the entire distribution

- We are dealing with 363 individuals assets and we ignore the correlation between them.
- From our calculation we can see that Skewness and Kurtosis are quite high

So we conclude that choosing a portfolio weights based only on the Sharpe ratios of the stocks in our dataset will not be a good choice.
